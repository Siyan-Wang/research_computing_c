```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline 
import pandas as pd
```

```python
topodata = pd.read_csv('/Users/admin/desktop/research_computing/Assignment9/AASZ_topo.txt', sep="\t", header=1)
topodata.columns = ["long", "lat", "distance_m", "elevation_m"]
topodata.head()
```

```python
topodata.distance_m=topodata.distance_m*1000
topodata.head()
```

```python
topodata['slope']=np.arctan(topodata.elevation_m.diff()/topodata.distance_m.diff())
topodata.head()
```

```python
f, ax = plt.subplots(2,1,figsize=(10,8))
# fig, ax =plt.figure(2,1,figsize=(10,8))
ax[0].plot(topodata.distance_m,topodata.elevation_m)
ax[0].set_xlabel('distance_m',size=13)
ax[0].set_ylabel('elevation_m',size=13)
ax[1].plot(topodata.distance_m,topodata.slope)
ax[1].set_xlabel('distance_m',size=13)
ax[1].set_ylabel('slope',size=13)
plt.tight_layout()
```

```python
from matplotlib.backends.backend_pdf import PdfPages
pp = PdfPages('/Users/admin/desktop/research_computing/Assignment9/AASZ_topo_slope.pdf')
pp.savefig(f)
pp.close()
```
